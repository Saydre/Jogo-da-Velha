```python
#importação das bibliotecas
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
def mostrar(imagem):
    fig = plt.gcf()
    fig.set_size_inches(18,6)
    plt.imshow(imagem, cmap='gray')
    plt.axis('off')
    plt.show()
```

```python
img = cv2.imread('/content/gdrive/MyDrive/imagens/paisagem03.jpg')
plt.imshow(img)
```

```python
original = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(original)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(gray)
```

```python
plt.imshow(gray, cmap='gray')
```

```python
limiar = 180 #valor < 180 = 0 e valor >= 180 = 255
```

```python
val, thresh = cv2.threshold(gray, limiar, 255, cv2.THRESH_BINARY)
```

```python
#Valor da limiarização
print(f'O valor da limiariazação é {val}')
```

```python
#Imagem limiarizada
thresh
```

```python
# Mostrar matriz completa
np.set_printoptions(threshold=np.inf)
```

```python
_, thresh = cv2.threshold(gray, limiar, 255, cv2.THRESH_BINARY)
mostrar(thresh)
```

```python
_, thresh = cv2.threshold(gray, limiar, 255, cv2.THRESH_BINARY_INV)
mostrar(thresh)
```

```python
_, thresh = cv2.threshold(gray, limiar, 255, cv2.THRESH_TRUNC)
mostrar(thresh)
```

```python
_, thresh = cv2.threshold(gray, limiar, 255, cv2.THRESH_TOZERO)
mostrar(thresh)
```

```python
_, thresh = cv2.threshold(gray, limiar, 255, cv2.THRESH_TOZERO_INV)
mostrar(thresh)
```
